```python
%%manim -qm PointOnSphere

class PointOnSphere(ThreeDScene):
    def construct(self):
        axes = ThreeDAxes()
        self.set_camera_orientation(phi=75 * DEGREES, theta=30 * DEGREES)
        
        # Create the sphere with a larger radius, say 2
        sphere = Sphere(radius=2, color=BLUE).set_opacity(0.5)
        
        # Create the path
        def path(t):
            return np.array([2 * np.cos(t) * np.cos(t), 2 * np.sin(t) * np.cos(t), 2 * np.sin(t)])
        
        path_points = [path(t) for t in np.linspace(0, np.pi/4, 100)]
        path_line = Line(path_points[0], path_points[1], color=RED, stroke_width=5)
        for i in range(1, len(path_points) - 1):
            path_line.add_line_to(path_points[i + 1])
        
        # Create arrows and labels
        arrows = VGroup()
        labels = VGroup()
        for t in np.linspace(0, np.pi/4, 5):
            point = path(t)
            arrow = Arrow3D(ORIGIN, point, color=YELLOW)
            lat_label = MathTex(f"\\text{{Lat: }} {int(180 * t / np.pi)}^\\circ").next_to(point, RIGHT).scale(0.2)
            lon_label = MathTex(f"\\text{{Lon: }} {int(180 * t / np.pi)}^\\circ").next_to(point, UP).scale(0.2)
            arrows.add(arrow)
            labels.add(lat_label, lon_label)
        
        # Create dashed lines
        dashed_lines = VGroup()
        for t in np.linspace(0, np.pi/4, 5):
            point = path(t)
            dashed_line = DashedLine(ORIGIN, point, color=WHITE, dash_length=0.05).set_stroke(width=2)
            dashed_lines.add(dashed_line)
        
        self.add(axes, sphere, path_line, arrows, labels, dashed_lines)

        self.begin_ambient_camera_rotation(rate=0.3)
        self.wait(15)
        self.stop_ambient_camera_rotation()
```

```python
%%manim -qm SinSurface

class SinSurface(ThreeDScene):
    def construct(self):
        axes = ThreeDAxes()
        self.set_camera_orientation(phi=75 * DEGREES, theta=30 * DEGREES)
        
        def func(x, y):
            return np.sin(x + y**2)
        
        # Create the main surface
        surface = Surface(
            lambda u, v: axes.c2p(u, v, func(u, v)),
            u_range=[-2, 2],
            v_range=[-2, 2],
            resolution=(50, 50),
            color=BLUE
        ).set_opacity(0.5)
        
        # Create the bottom filling surface with condition
        def filling_func(u, v):
            if u**2 + v**2 < 4:
                return axes.c2p(u, v, 0)
            else:
                return axes.c2p(u, v, func(u, v))
        
        filling = Surface(
            filling_func,
            u_range=[-2, 2],
            v_range=[-2, 2],
            resolution=(50, 50),
            color=RED
        ).set_opacity(0.4)
        
        # Add everything to the scene
        self.add(axes, surface, filling)
        self.wait(5)
```

```python
%%manim -qm BesselPlot

from scipy.special import jn

class BesselPlot(Scene):
    def construct(self):
        axes = Axes(
            x_range=[0, 10.5], 
            y_range=[-0.5, 0.7], 
            axis_config={"color": BLUE},
        )
        self.play(Create(axes))

        colors = [YELLOW, ORANGE, RED, GREEN]

        # This function returns the Bessel function values
        def bessel_function(n, x):
            return jn(n, x)

        # Plotting the Bessel functions and animating them
        for n in range(4):
            graph = axes.plot(lambda x: bessel_function(n, x), color=colors[n])
            fill = axes.get_area(graph, color=colors[n], opacity=0.3)
            self.play(Create(graph), FadeIn(fill), run_time=2)
        
        self.wait(1)
```

```python
%%manim -qm ParaSurface
config.media_width = "50%"

class ParaSurface(ThreeDScene):
    def func(self, u, v):
        # parametrize u,v onto a circle with r=2,
        # u as angle, v as radius
        x = v*np.cos(u)
        y = v*np.sin(u)
        return np.array([x, y, np.sin(x+y**2)])

    def construct(self):
        axes = ThreeDAxes(
            x_range=[-2.5,2.5], x_length=8,
            y_range=[-2.5,2.5], y_length=8,
        )
        surface = Surface(
            lambda u, v: axes.c2p(*self.func(u, v)),
            u_range=[-PI, PI],
            v_range=[0, 2],
            resolution=16,
        )
        self.set_camera_orientation(theta=70 * DEGREES, phi=75 * DEGREES)
        self.add(axes, surface)
```

```python
%%manim -qm Plot3DSine1
class Plot3DSine1(ThreeDScene):
    def construct(self):
        axes = ThreeDAxes(
            x_range=[-2, 2],
            y_range=[-2, 2],
            z_range=[-1.5, 1.5],
            x_length=4,
            y_length=4,
            z_length=3,
        )

        def func(x, y):
            return np.sin(x + y ** 2)

        surface = Surface(
            lambda u, v: axes.c2p(u, v, func(u, v)),
            u_range=[-2, 2],
            v_range=[-2, 2],
            resolution=[50, 50],
        ).set_opacity(0.4)

        def condition(u, v):
            return u ** 2 + v ** 2 < 4

        # Creating a surface for the filled region
        filled_region = Surface(
            lambda u, v: axes.c2p(u, v, 0),
            u_range=[-2, 2],
            v_range=[-2, 2],
            resolution=[50, 50],
            checkerboard_colors=[RED]
        ).set_opacity(0.4)

        # Using a mask to cut the filled region according to the condition
        mask = Surface(
            lambda u, v: axes.c2p(u, v, 0) if condition(u, v) else axes.c2p(u, v, -3),
            u_range=[-2, 2],
            v_range=[-2, 2],
            resolution=[50, 50],
            checkerboard_colors=[WHITE, WHITE]
        )
        
        masked_fill = Intersection(filled_region, mask)

        self.add(axes, masked_fill, surface)
```

```python
%%manim -qm Plot3DSurface

class Plot3DSurface(ThreeDScene):
    def construct(self):
        axes = ThreeDAxes()
        self.set_camera_orientation(phi=75 * DEGREES, theta=30 * DEGREES)
        
        def func(x, y):
            return np.sin(x + y**2)

        # Create the surface with masking for the circle
        surface_points = []
        for u in np.linspace(-2, 2, 30):
            for v in np.linspace(-2, 2, 30):
                if u**2 + v**2 < 4:
                    surface_points.append(axes.c2p(u, v, func(u, v)))
                else:
                    surface_points.append(axes.c2p(u, v, 0))

        surface = Surface(
            lambda u, v: surface_points[int(u*29)*30 + int(v*29)],
            u_range=[0, 1], v_range=[0, 1],
            checkerboard_colors=[BLUE, BLUE_D]
        )

        # Fill below the surface
        fill = Surface(
            lambda u, v: axes.c2p(u*4-2, v*4-2, 0),
            u_range=[0, 1], v_range=[0, 1],
            color=RED,
            fill_opacity=1
        ).set_shade_in_3d(True)

        self.add(axes, surface, fill)
```

```python
%%manim -qm PlotFunction

class PlotFunction(Scene):
    def construct(self):
        axes = Axes(
            x_range=[0, 2],
            y_range=[-2, 2],
            axis_config={"color": BLUE},
        )
        
        # Function definition that avoids NaNs
        def func(x):
            return x * np.sin(1 / x) if x != 0 else 0

        graph = axes.plot(lambda x: func(x), x_range=[0.01, 0.1], color=WHITE)

        vertical_line = axes.get_vertical_line(
            axes.c2p(0.1, 0),
            color=YELLOW,
        )

        def update_graph(graph, alpha):
            r_value = interpolate(0.1, 2, alpha)
            graph.become(
                axes.plot(lambda x: func(x), x_range=[0.01, r_value], color=WHITE)
            )

        def update_line(vertical_line):
            r_value = axes.point_to_coords(graph.get_end())[0]
            vertical_line.become(
                axes.get_vertical_line(axes.c2p(r_value, 0))
            )

        self.play(Create(axes), Create(graph), Create(vertical_line))
        self.play(
            UpdateFromAlphaFunc(graph, update_graph),
            UpdateFromFunc(vertical_line, update_line),
            run_time=5,
        )
```
